```python
import numpy as np
import mdtraj as md
from matplotlib import pyplot as plt
```

```python
colvar_file = open('../Inputs/A_lugano/COLVAR', 'r')

time, phi = [], []
for line in colvar_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        if (float(data[0]) % 10 == 0):
            time.append(float(data[0]) / 1000)
            phi.append(float(data[1]))

#print(time[0], phi[0])
#print(time[100], phi[100])
```

```python
plt.plot(time, phi, 'o')
plt.title("Phi over timestep of metadynamics simulation A")
plt.xlabel("Time (ns)")
plt.ylabel("Phi (Radians)")
plt.savefig("../Figures/phi_over_time.png")
plt.show()
```

```python
hills_file = open('../Inputs/A_lugano/HILLS', 'r')

time, height = [], []
for line in hills_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        time.append(float(data[0]) / 1000)
        height.append(float(data[3]))
#print(time[0], height[0])
#print(time[100], height[100])
```

```python
plt.plot(time, height, 'o')
plt.title("Gaussian Hill Height over timestep of metadynamics simulation A")
plt.xlabel("Time (ns)")
plt.ylabel("Gaussian Hill Height")
plt.savefig("../Figures/gaussians_over_time.png")
plt.show()
```

```python
grid_file = open('../Inputs/A_lugano/gaussians.grid.dat', 'r')

phi, eng = [], []
for line in grid_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi.append(float(data[0]))
        eng.append(-float(data[1]))
        
plt.plot(phi, eng)
plt.title("Free Energy Accumulations by Phi Angle")
plt.xlabel("Phi (radians)")
plt.savefig("../Figures/free_eng_by_phi.png")
plt.show()
```

```python
sum_hills_file = open('../Inputs/A_lugano/fes.dat', 'r')

phi, eng = [], []
for line in sum_hills_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi.append(float(data[0]))
        eng.append(float(data[1]))

plt.plot(phi, eng)
plt.title("Free Energy by Phi Angle")
plt.xlabel("Phi (radians)")
plt.savefig("../Figures/sum_hills_phi.png")
plt.show()
```

```python
sum_hills_file_1 = open('../Inputs/A_lugano/fes_1.dat', 'r')
sum_hills_file_3 = open('../Inputs/A_lugano/fes_3.dat', 'r')
sum_hills_file_10 = open('../Inputs/A_lugano/fes_10.dat', 'r')
sum_hills_file_20 = open('../Inputs/A_lugano/fes_20.dat', 'r')
sum_hills_file_50 = open('../Inputs/A_lugano/fes_50.dat', 'r')
sum_hills_file_100 = open('../Inputs/A_lugano/fes_100.dat', 'r')

phi1, eng1 = [], []
phi3, eng3 = [], []
phi10, eng10 = [], []
phi20, eng20 = [], []
phi50, eng50 = [], []
phi100, eng100 = [], []
for line in sum_hills_file_1.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi1.append(float(data[0]))
        eng1.append(float(data[1]))
for line in sum_hills_file_3.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi3.append(float(data[0]))
        eng3.append(float(data[1]))
for line in sum_hills_file_10.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi10.append(float(data[0]))
        eng10.append(float(data[1]))
for line in sum_hills_file_20.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi20.append(float(data[0]))
        eng20.append(float(data[1]))
for line in sum_hills_file_50.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi50.append(float(data[0]))
        eng50.append(float(data[1]))
for line in sum_hills_file_100.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi100.append(float(data[0]))
        eng100.append(float(data[1]))


plt.plot(phi1, eng1, label="0.1 ns")
plt.plot(phi3, eng3, label="0.3 ns")
plt.plot(phi10, eng10, label="1 ns")
plt.plot(phi20, eng20, label="2 ns")
plt.plot(phi50, eng50, label="5 ns")
plt.plot(phi100, eng100, label="10 ns")
plt.title("Free Energy by Phi Angle")
plt.xlabel("Phi (radians)")
plt.legend()
plt.savefig("../Figures/sum_hills_phi_over_time.png")
plt.show()
```

```python
sum_hills_file_1 = open('../Inputs/A_2D_lugano/fes_1.dat', 'r')
sum_hills_file_3 = open('../Inputs/A_2D_lugano/fes_3.dat', 'r')
sum_hills_file_10 = open('../Inputs/A_2D_lugano/fes_10.dat', 'r')
sum_hills_file_20 = open('../Inputs/A_2D_lugano/fes_20.dat', 'r')
sum_hills_file_50 = open('../Inputs/A_2D_lugano/fes_50.dat', 'r')
sum_hills_file_100 = open('../Inputs/A_2D_lugano/fes_100.dat', 'r')

phi1, psi1, eng1 = [], [], []
phi3, psi3, eng3 = [], [], []
phi10, psi10, eng10 = [], [], []
phi20, psi20, eng20 = [], [], []
phi50, psi50, eng50 = [], [], []
phi100, psi100, eng100 = [], [], []
for line in sum_hills_file_1.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi1.append(float(data[0]))
        psi1.append(float(data[1]))
        eng1.append(float(data[2]))
for line in sum_hills_file_3.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi3.append(float(data[0]))
        psi3.append(float(data[1]))
        eng3.append(float(data[2]))
for line in sum_hills_file_10.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi10.append(float(data[0]))
        psi10.append(float(data[1]))
        eng10.append(float(data[2]))
for line in sum_hills_file_20.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi20.append(float(data[0]))
        psi20.append(float(data[1]))
        eng20.append(float(data[2]))
for line in sum_hills_file_50.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi50.append(float(data[0]))
        psi50.append(float(data[1]))
        eng50.append(float(data[2]))
for line in sum_hills_file_100.readlines():
    if line[0] != '#' and line[0] != '\n':
        data = line.strip().split()
        phi100.append(float(data[0]))
        psi100.append(float(data[1]))
        eng100.append(float(data[2]))

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi1, psi1, eng1)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("0.1 ns")
plt.savefig("../Figures/2D_0_1_ns.png")
plt.show()

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi3, psi3, eng3)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("0.3 ns")
plt.savefig("../Figures/2D_0_3_ns.png")
plt.show()

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi10, psi10, eng10)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("1 ns")
plt.savefig("../Figures/2D_1_ns.png")
plt.show()

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi20, psi20, eng20)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("2 ns")
plt.savefig("../Figures/2D_2_ns.png")
plt.show()

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi50, psi50, eng50)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("5 ns")
plt.savefig("../Figures/2D_5_ns.png")
plt.show()

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(phi100, psi100, eng100)
plt.xlabel("Phi")
plt.ylabel("Psi")
plt.title("10 ns")
plt.savefig("../Figures/2D_10_ns.png")
plt.show()
```

```python
grid_2D_file = open("../Inputs/A_2D_lugano/gaussians.grid.dat")
grid = []
vector = []
phi_done = False
psi_next = True
phis = []
psis = []
for line in grid_2D_file.readlines():
    if line[0] != '#':
        if line[0] == '\n':
            grid.append(vector)
            vector = []
            psi_next = True
        else:
            data = line.strip().split()
            if psi_next:
                psis.append(float(data[1]))
                psi_next = False
            if not phi_done:
                phis.append(float(data[0]))
            vector.append(float(data[2]))
plt.imshow(grid)
```
