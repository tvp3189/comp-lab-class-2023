```python
import numpy as np
import mdtraj as md
from matplotlib import pyplot as plt
```

```python
colvar_file = open('../Inputs/A_lugano/COLVAR', 'r')

time, phi = [], []
for line in colvar_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        if (float(data[0]) % 10 == 0):
            time.append(float(data[0]) / 1000)
            phi.append(float(data[1]))

#print(time[0], phi[0])
#print(time[100], phi[100])
```

```python
plt.plot(time, phi, 'o')
plt.title("Phi over timestep of metadynamics simulation A")
plt.xlabel("Time (ns)")
plt.ylabel("Phi (Radians)")
plt.show()
```

```python
hills_file = open('../Inputs/A_lugano/HILLS', 'r')

time, height = [], []
for line in hills_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        time.append(float(data[0]) / 1000)
        height.append(float(data[3]))
#print(time[0], height[0])
#print(time[100], height[100])
```

```python
plt.plot(time, height, 'o')
plt.title("Gaussian Hill Height over timestep of metadynamics simulation A")
plt.xlabel("Time (ns)")
plt.ylabel("Gaussian Hill Height")
plt.show()
```

```python
grid_file = open('../Inputs/A_lugano/gaussians.grid.dat', 'r')

phi, eng = [], []
for line in grid_file.readlines():
    if line[0] != '#':
        data = line.strip().split()
        phi.append(float(data[0]))
        eng.append(-float(data[1]))
        
plt.plot(phi, eng)
plt.title("Free Energy Accumulations by Phi Angle")
plt.xlabel("Phi (radians)")
plt.show()
```
