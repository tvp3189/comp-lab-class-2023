```python
import mdtraj as md
import numpy as np
from matplotlib import pyplot as plt
```

```python
# import data file from simulation
energy_data_file = open('../Data/1hz3_T310.run.25000000.energy.xvg', 'r')
```

```python
# read in data ignoring comment lines
energy_data_lines = []
for line in energy_data_file:
    if line[0] != '#' and line[0] != '@':
        energy_data_lines.append(line)
```

```python
time, pot_eng, kin_eng, tot_eng, temp, pres, vol = [], [], [], [], [], [], []
for line in energy_data_lines:
    data = line.strip().split()
    #if len(data) != 7:
    #   print(len(time))
    time.append(float(data[0]) / 1000) # Time is converted from ps to ns
    pot_eng.append(float(data[1]))
    kin_eng.append(float(data[2]))
    tot_eng.append(float(data[3]))
    temp.append(float(data[4]))
    pres.append(float(data[5]))
    vol.append(float(data[6]))
```

```python
plt.title('Potential')
plt.xlabel('Time (ns)')
plt.ylabel('Potential')
plt.plot(time, pot_eng)
```

```python
plt.title('Kinetic Energy over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Kinetic Energy')
plt.plot(time, kin_eng)
```

```python
plt.title('Total Energy over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Total Energy')
plt.plot(time, tot_eng)
```

```python
plt.title('Temperature over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Temperature (K)')
plt.plot(time, temp)
```

```python
plt.title('Volume over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Volume')
plt.plot(time, vol)
```

```python
plt.title('Pressure over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Pressure')
plt.plot(time, pres)
```

```python
plt.title('Temperature vs. Kinetic Energy')
plt.xlabel('Kinetic Energy')
plt.ylabel('Temperature (K)')
plt.plot(kin_eng, temp, 'r')
```
