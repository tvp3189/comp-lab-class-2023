```python
import mdtraj as md
import numpy as np
from matplotlib import pyplot as plt
```

```python
traj = md.load_trr("../Inputs/adp_T300_initial/backups2/adp_T300.trr", top="../Inputs/adp_T300_initial/backups2/adp_T300.gro")
```

```python
(i, phi) = md.compute_phi(traj)
(j, psi) = md.compute_psi(traj)

plt.plot(phi, 'o')
plt.title("Phi over Time of Initial Analysis")
plt.xlabel("Time")
plt.ylabel("Phi")
plt.show()

plt.plot(psi, 'o')
plt.title("Psi over Time of Initial Analysis")
plt.xlabel("Time")
plt.ylabel("Psi")
plt.show()
```

```python
hist, __, __ = np.histogram2d(phi.ravel(), psi.ravel(), bins=50)
def Fn(i):
    if i == 0:
        return -1e10
    return -np.log(i)

for i in range(len(hist)):
    for j in range(len(hist[i])):
        hist[i][j] = Fn(hist[i][j])

plt.pcolormesh(hist.T, cmap="viridis", vmin=-10, vmax=0)
plt.title("Free Energy Surface of Alanine Dipeptide")
plt.show()

#plt.hist2d(phi.ravel(), psi.ravel(), bins=50)
#plt.title("Ramachandran Plot of T300")
#plt.xlabel("Phi")
#plt.ylabel("Psi")
#plt.colorbar()
#plt.show()
```

```python
traj2 = md.load_trr("../Inputs/T300/adp_exchange3temps.trr", top="../Inputs/T300/adp_exchange3temps.gro")
```

```python
(i, phi_2) = md.compute_phi(traj2)
(j, psi_2) = md.compute_psi(traj2)


hist, __, __ = np.histogram2d(phi_2.ravel(), psi_2.ravel(), bins=50)
def Fn(i):
    if i == 0:
        return -1e10
    return -np.log(i)

for i in range(len(hist)):
    for j in range(len(hist[i])):
        hist[i][j] = Fn(hist[i][j])

plt.pcolormesh(hist.T, cmap="viridis", vmin=-10, vmax=0)
plt.title("Second Free Energy Surface of Alanine Dipeptide")
plt.show()
```

```python
temp = open("../Inputs/T300/replica_temp.xvg", "r")
time, temp1, temp2, temp3 = [], [], [], []
#print(index.readlines()[4000])

for line in temp.readlines():
    datapt = line.strip().split()
    if (float(datapt[0]) % 100 == 0):
        time.append(float(datapt[0]) / 1000)
        temp1.append(int(datapt[1]) + 1)
        temp2.append(int(datapt[2]) + 1)
        temp3.append(int(datapt[3]) + 1)
```

```python
plt.title("Temperature Replica Exchange of Each System")
print(temp1)
plt.plot(time, temp1, label="T300")
plt.plot(time, temp2, label="T363")
plt.plot(time, temp3, label="T440")
plt.xlabel("Time (ns)")
plt.legend()
plt.show()
```
