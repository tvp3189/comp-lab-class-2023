```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def open_xvg(filename):
    data = []
    raw_data = open(filename, 'r')
    for line in raw_data:
        if line[0] != '#' and line[0] != '@':
            datapt = cast_float(line.strip().split())
            data.append(datapt)
    return data
                        
def cast_float(raw_datapt):
    datapt = []
    for item in raw_datapt:
        datapt.append(float(item))
    return datapt

def plot_xvg(filename):
    datapts = open_xvg(filename)
    time, measure = [], []
    for pt in datapts:
        time.append(pt[0])
        measure.append(pt[1])
    return time, measure
```

```python
# Plot Potential Energy during Energy Minimization
time, potential = plot_xvg('potential.xvg')

plt.xlabel("Energy Minimization Step")
plt.ylabel("Potential Energy (kJ/mol)")
plt.title("Energy Minimization of 1AKI by Steepest Descent")

plt.plot(time, potential)
plt.show()
plt.savefig('../Figures/potential.png')
```

```python
# Plot Temperature during NVT Equilibration
time, eq_temp = plot_xvg('eq_temperature.xvg')

plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("Temperature of NVT Equilibration of 1AKI")

plt.plot(time, eq_temp)
plt.show()
plt.savefig('../Figures/eq_temp.png')
```

```python
# Plot Pressure during NPT Equilibration
time, eq_pres = plot_xvg('npteq_pressure.xvg')

plt.xlabel("Time (ps)")
plt.ylabel("Pressure (bar)")
plt.title("Pressure of NPT Equilibration of 1AKI")

plt.plot(time, eq_pres)
plt.show()
plt.savefig('../Figures/eq_pres.png')
```

```python
# Plot Density during NPT Equilibration
time, eq_dens = plot_xvg('npteq_density.xvg')

plt.xlabel("Time (ps)")
plt.ylabel("Density (kg m^-3)")
plt.title("Density of NPT Equilibration of 1AKI")

plt.plot(time, eq_dens)
plt.show()
plt.savefig('../Figures/eq_dens.png')
```

```python
# Plot RMSD during Production
# Compare to initial struc and crystal struc
time, rmsd_initial_struct = plot_xvg('rmsd.xvg')
time, rmsd_crystal_struct = plot_xvg('rmsd_xtal.xvg')


plt.xlabel("Time (ns)")
plt.ylabel("RMSD (nm)")
plt.title("RMSD of 1AKI")

plt.plot(time, rmsd_initial_struct, label='Ref: Equilibrated')
plt.plot(time, rmsd_crystal_struct, label='Ref: Crystal')
plt.legend()
plt.show()
plt.savefig('../Figures/rmsd.png')
```

```python
# Plot Rg during Production
time, rg = plot_xvg('gyrate.xvg')

plt.xlabel("Time (ns)")
plt.ylabel("Rg (nm)")
plt.title("Radius of gyration of 1AKI")

plt.plot(time, rg)
plt.show()
plt.savefig('../Figures/')
```
