```python
import psi4
psi4.set_output_file("output.dat", False)
psi4.set_memory("500 MB")
```

```python
%%time
ar2_geometry = """
Ar
--
Ar 1 {0}
"""

Rvals = [3.5, 3.6, 3.7, 3.8, 3.9, 
         4.0, 4.1, 4.2, 4.3, 4.4, 
         4.5, 5.0, 5.5, 6.0]

psi4.set_options({'freeze_core': True})

ecp = {}

for R in Rvals:
    ar2 = psi4.geometry(ar2_geometry.format(R))
    ecp[R] = psi4.energy('ccsd(t)/aug-cc-pvdz', bsse_type='cp', molecule=ar2)
    
# Prints to screen
print("CP-corrected CCSD(T)/aug-cc-pVDZ Interaction Energies\n\n")
print("          R [Ang]                 E_int [kcal/mol]       ")
print("---------------------------------------------------------")
for R in Rvals:
    e = ecp[R] * psi4.constants.hartree2kcalmol
    print("            {:3.1f}                        {:1.6f}".format(R, e))

# Prints to output.dat
psi4.core.print_out("CP-corrected CCSD(T)/aug-cc-pVDZ Interaction Energies\n\n")
psi4.core.print_out("          R [Ang]                 E_int [kcal/mol]       \n")
psi4.core.print_out("---------------------------------------------------------\n")

energy = []
for R in Rvals:
    e = ecp[R] * psi4.constants.hartree2kcalmol
    energy.append(e)
    psi4.core.print_out("            {:3.1f}                        {:1.6f}\n".format(R, e))
```

```python
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
def lj(r,sigma,eps):
    return 4*eps*( (sigma/r)**12-(sigma/r)**6  )


param, param_cov = curve_fit(lj, Rvals, energy)

x = np.linspace(3.5, 6.0)
y = lj(x, param[0], param[1])
```

```python
#print(Rvals, energy)
#print(x,y)
plt.plot()
plt.plot(Rvals, energy, 'o')
plt.plot(x, y, '--')
plt.title("Argon Dimer fit to Leonard-Jones Potential")
plt.savefig("../Figures/argon_LJ.png")
plt.show()
```
