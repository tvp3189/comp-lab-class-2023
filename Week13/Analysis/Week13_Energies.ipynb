```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
npt_file = open('../Outputs/NPT-1.ener', 'r')
nvt_file = open('../Outputs/NVT-1.ener', 'r')
```

```python
npt_time, npt_KE, npt_U = [], [], []
for line in npt_file.readlines():
    if line[0] == '#':
        continue
    data = line.strip().split()
    step_num = int(data[0])
    time = float(data[1])
    KE = float(data[2])
    temp = float(data[3])
    U = float(data[4])
    
    npt_time.append(time)
    npt_KE.append(KE)
    npt_U.append(U)
    
nvt_time, nvt_KE, nvt_U = [], [], []
for line in nvt_file.readlines():
    if line[0] == '#':
        continue
    data = line.strip().split()
    step_num = int(data[0])
    time = float(data[1])
    KE = float(data[2])
    temp = float(data[3])
    U = float(data[4])
    
    nvt_time.append(time)
    nvt_KE.append(KE)
    nvt_U.append(U)
```

```python
plt.plot(npt_time, npt_KE)
plt.title('NPT Equilibration Kinetic Energy')
plt.plot()
plt.show()

plt.plot(npt_time, npt_U)
plt.title('NPT Equilibration Potential Energy')
plt.plot()
plt.savefig('../Figures/npt_U.png')
plt.show()

plt.plot(nvt_time, nvt_KE)
plt.title('NVT Equilibration Kinetic Energy')
plt.plot()
plt.show()

plt.plot(nvt_time, nvt_U)
plt.title('NVT Equilibration Potential Energy')
plt.plot()
plt.savefig('../Figures/nvt_U.png')
plt.show()
```
