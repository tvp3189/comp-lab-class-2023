```python
from matplotlib import pyplot as plt
```

```python
std_rxn_file = open('../Outputs/monitor/MONITOR-COLVAR.metadynLog')

std_time, std_rxn = [], []
for line in std_rxn_file.readlines():
    data = line.strip().split()
    time = float(data[0])
    rxn_coord = float(data[1])
    
    std_time.append(time)
    std_rxn.append(rxn_coord)

plt.plot(std_time, std_rxn)
plt.title('Reaction Coordinate of Unbiased Run')
plt.xlabel('Time')
plt.ylabel('r(CO) - r(CC)')
plt.savefig('../Figures/unbiased_rxn_coord.png')
plt.show()
```
